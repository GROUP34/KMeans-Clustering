```python
x_train = data.iloc[:,:4]
```

```python
x_scaled = scale(x_train)
```

```python
# Kmeans
wcss = [] #within cluster sum of squares
num_of_clusters = range(1,10)

for i in num_of_clusters:
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(num_of_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Num of Clusters')
plt.ylabel('Within-Cluster Sum of squares')
plt.show
```

```python
#plot clusters
kmeans = KMeans(3)
kmeans.fit(x_scaled)
data_with_clusters = x_train.copy()
data_with_clusters['Clusters'] = kmeans.fit_predict(x_scaled)
```
